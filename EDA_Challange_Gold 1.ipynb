```python
def alay_word_ratio(data_df, new_kamusalay_df):
  """Calculates and prints the ratio and percentage of alay words in the data.

  Args:
    data_df: DataFrame containing the text data.
    new_kamusalay_df: DataFrame containing a list of alay words.
  """

  if data_df is None or new_kamusalay_df is None:
    return

  alay_words = set(new_kamusalay_df['anakjakartaasikasik'].astype(str).tolist())
  alay_count = 0
  total_word_count = 0

  for tweet in data_df['Tweet'].astype(str).tolist():
    words = tweet.lower().split()
    total_word_count += len(words)
    for word in words:
      if word in alay_words:
        alay_count += 1

  if total_word_count > 0:
    percentage = (alay_count / total_word_count) * 100

    print(f"Kata alay : Data seluruh = {alay_count} : {total_word_count}")
    print(f"Persentase: {percentage:.2f}%")
  else:
    print("Tidak ada data untuk diproses.")

alay_word_ratio(data_df, new_kamusalay_df)
```

```python
def abusive_word_ratio(data_df, abusive_df):
  """Calculates and prints the ratio and percentage of abusive words in the data.

  Args:
    data_df: DataFrame containing the text data.
    abusive_df: DataFrame containing a list of abusive words.
  """
  if data_df is None or abusive_df is None:
    return

  abusive_words = set(abusive_df['Abusive'].astype(str).tolist())
  abusive_count = 0
  total_word_count = 0

  for tweet in data_df['Tweet'].astype(str).tolist():
    words = tweet.lower().split()
    total_word_count += len(words)
    for word in words:
      if word in abusive_words:
        abusive_count += 1

  if total_word_count > 0:
    percentage = (abusive_count / total_word_count) * 100
    print(f"Kata abusive : Data seluruh = {abusive_count} : {total_word_count}")
    print(f"Persentase: {percentage:.2f}%")
  else:
    print("Tidak ada data untuk diproses.")

abusive_word_ratio(data_df, abusive_df)
```
